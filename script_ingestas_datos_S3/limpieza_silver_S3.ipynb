```python
# =====================================================================
# SCRIPT DE LIMPIEZA - Borra contenido de silver/ en S3 (Databricks)
# Mantiene la estructura de carpetas vacías para las 4 clases.
# =====================================================================
from rich import print

BUCKET_OUTPUT = 'shazam-proyecto-ecg'
S3_BASE = f"s3://{BUCKET_OUTPUT}/silver"

CARPETAS = [
    "silver/clase_0_normal/",
    "silver/clase_1_afib/",
    "silver/clase_2_pvc/",
    "silver/clase_3_lbbb/",
]

print("[bold cyan]--- LIMPIEZA DE SILVER EN S3 ---[/bold cyan]")
print(f"[white]Bucket: {BUCKET_OUTPUT}[/white]")
print(f"[white]Prefijo base: silver/[/white]\n")

# 1. Eliminar todo el contenido recursivamente
for carpeta in CARPETAS:
    ruta = f"s3://{BUCKET_OUTPUT}/{carpeta}"
    try:
        archivos = dbutils.fs.ls(ruta)
        n_archivos = len(archivos)
        if n_archivos > 0:
            dbutils.fs.rm(ruta, recurse=True)
            print(f"[bold green][OK][/bold green] Eliminados {n_archivos} objetos en: {carpeta}")
        else:
            print(f"[yellow][SKIP][/yellow] Ya vacía: {carpeta}")
    except Exception:
        print(f"[yellow][SKIP][/yellow] No existe aún: {carpeta}")

# 2. Recrear carpetas vacías (mantener estructura)
print("\n[yellow]Recreando estructura de carpetas...[/yellow]")
for carpeta in CARPETAS:
    ruta = f"s3://{BUCKET_OUTPUT}/{carpeta}"
    dbutils.fs.mkdirs(ruta)
    print(f"[bold green][OK][/bold green] Carpeta creada: {carpeta}")

print("\n[bold green]--- LIMPIEZA COMPLETADA ---[/bold green]")
print("[white]Estructura silver/ lista para nuevas ingestas.[/white]")
```
